```python
import pandas as pd
import numpy as np
```

```python
rela = pd.read_csv("data1.csv")
allp = pd.read_csv("AllProducts_June26.csv")
```

```python
for i in range(allp.shape[0]):
    allp.Name[i] = allp.Name[i].strip()
```

```python
c = 0
for i in rela[rela["Relation"] == "Dad"].Name:
    if i in allp.Name:
        c = c+1
c
```

```python
res_merge = allp.append(rela)
res_merge = res_merge.reset_index()
for i in range(res_merge.shape[0]):
    if type(res_merge['Regular price'][i]) == str:
        res_merge.loc[i, 'Regular price'] = int(''.join(res_merge.loc[i, 'Regular price'][1:-3].split(',')))
```

```python
res_merge.drop(columns = ['index'],inplace = True)
```

```python
res_merge.to_csv("Merged.csv",index = False)
```
